**Recreate the plot below. Use tips dataset. Hint: pay attention to x and y labels.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!

plt.figure(figsize=(5, 3))  # Set the figure size
sns.boxplot(x="day", y="tip", data=tips, hue="sex",palette=["blue", "green"])
plt.legend(loc='upper right',title="sex")
```

**Recreate the plot below. Use tips dataset. Hint: pay attention to x and y labels.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
plt.figure(figsize=(4, 5)) 
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

**Recreate the distplot given below. Use titanic dataset. Hint: pay attention to x and y labels.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
plt.figure(figsize=(5, 3))

sns.distplot(titanic['fare'],kde=False,bins=30,color='lightgreen')
plt.show()


```

**Recreate the boxplot below using titanic dataset. Hint: pay attention to x and y labels.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
plt.figure(figsize=(5, 4))  # Set the figure size
sns.boxplot(x="class", y="age", data=titanic, hue="alive", palette="pastel")
plt.legend(loc='upper right',title="alive")
```

**Recreate the plot below using titanic dataset. Hint: pay attention to x and y labels.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!

sns.jointplot(x='fare',y='age',data=titanic)
plt.show()
```

**Recreate the plot below. Use titanic dataset. ** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
plt.figure(figsize=(4, 3)) 
sns.boxplot(x="class", y="age", data=titanic,palette=["blue", "lightgreen",'orange'])
```

**Can you plot how many male and female were on titanic? Separate them based on their class.**

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
plt.figure(figsize=(4, 3))
sns.countplot(data=titanic, x='sex', hue='class', palette={'First': 'blue', 'Second': 'green', 'Third': 'red'})

# Titles and Labels

plt.legend(title='class')

# Show Plot
plt.tight_layout()
plt.show()
```

**Recreate the plot below. Use titanic dataset. Hint: Matrix plots** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!

titanic2 = titanic[[ 'survived',
    'pclass',
    'age',
    'sibsp',
    'parch',
    'fare',
    'adult_male',
    'alone']]

tc=titanic2.corr()
tc
sns.heatmap(tc,cmap='coolwarm', annot=True, vmin=-0.8, vmax=0.8,cbar_kws={'ticks': [-0.8, -0.4, 0, 0.4, 0.8]})
plt.title('titanic.corr()') 
plt.show()

```
